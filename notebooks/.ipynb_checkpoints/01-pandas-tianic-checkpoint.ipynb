```python
path = '../data/titanic.csv'
df = pd.read_csv(path)
df
```

```python
clean_age1 = df['Age'].fillna(df['Age'].mean())
clean_age1.value_counts()
```

```python
clean_age2=df['Age'].dropna()
clean_age2
```

```python
age_counts = clean_age2.value_counts()
age_counts[:10]

```

```python
import seaborn as sns
ageset = age_counts[:10]
ageset
```

```python
sns.barplot(x=ageset.index, y=ageset.values)
```

```python
cframe=df[df.Age.notnull() & df.Sex.notnull()]
cframe
```

```python
by_sex_age = cframe.groupby(['Age', 'Sex'])
by_sex_age.size()
```

```python
agg_counts = by_sex_age.size().unstack().fillna(0)
agg_counts
```

```python
indexer = agg_counts.sum(1)
indexer
```

```python
# Use to sort in ascending order
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

```python
indexer[-10:]
```

```python
count_subset = agg_counts.take(indexer[-10:])
count_subset=count_subset.tail(10)
count_subset
```

```python
agg_counts.sum(1).nlargest(10)
```

```python
# Rearrange the data for plotting
stack_subset = count_subset.stack()
stack_subset
```

```python
stack_subset.name = 'total'
stack_subset = stack_subset.reset_index()
stack_subset
```

```python
sns.barplot(x='total', y='Age', hue='Sex',  data=stack_subset)
```

```python
def norm_total(group):
    group['normed_total'] = group.total / group.total.sum()
    return group

results = stack_subset.groupby('Age').apply(norm_total)
results
```

```python
sns.barplot(x='normed_total', y='Age', hue='Sex',  data=results)



```
