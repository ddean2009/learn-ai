## 普通Lasso和多任务Lasso的比较

```python
# Author: Alexandre Gramfort <alexandre.gramfort@inria.fr>
# License: BSD 3 clause

import matplotlib.pyplot as plt
import numpy as np

from sklearn.linear_model import MultiTaskLasso, Lasso
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
rng = np.random.RandomState(42)

# Generate some 2D coefficients with sine waves with random frequency and phase
n_samples, n_features, n_tasks = 100, 30, 40
n_relevant_features = 29
# 创建全为0的ndarray: 40 * 30
coef = np.zeros((n_tasks, n_features))
# linspace用来创建等差数列，在0和2* np.pi之间创建40个值
times = np.linspace(0, 2 * np.pi, n_tasks)
coef,times
```

```python
for k in range(n_relevant_features):
    # 构造系列的coef矩阵
    coef[:, k] = np.cos((1. + rng.randn(1)) * times + 3 * rng.randn(1))
# X 是100 * 30 的矩阵
X = rng.randn(n_samples, n_features)
# np.dot矩阵乘法得到100 * 40的矩阵
Y = np.dot(X, coef.T) + rng.randn(n_samples, n_tasks)
# Y.T是40 * 100的矩阵
coef_lasso_ = np.array([Lasso(alpha=0.5).fit(X, y).coef_ for y in Y.T])
coef_multi_task_lasso_ = MultiTaskLasso(alpha=1).fit(X, Y).coef_
```

```python
#test = np.ones((n_tasks, n_features))
# spy在所有坐标都非0的情况下会画出一张白图
#plt.spy(test)

# Plot support and time series
fig = plt.figure(figsize=(8, 5))
plt.subplot(1, 2, 1)
# 画图，0用白色表示，非0用黑色表示
plt.spy(coef_lasso_)
plt.xlabel('特征值')
plt.ylabel('Task')
plt.text(10, 5, 'Lasso')
plt.subplot(1, 2, 2)
plt.spy(coef_multi_task_lasso_)
plt.xlabel('特征值')
plt.ylabel('Task')
plt.text(10, 5, '多任务Lasso')
fig.suptitle('非零系数地址')

feature_to_plot = 0
plt.figure()
lw = 2
# x轴是index，从0到40， Y轴是coef的第一列
plt.plot(coef[:, feature_to_plot], color='seagreen', linewidth=lw,
         label='真实值')
plt.plot(coef_lasso_[:, feature_to_plot], color='cornflowerblue', linewidth=lw,
         label='Lasso')
plt.plot(coef_multi_task_lasso_[:, feature_to_plot], color='gold', linewidth=lw,
         label='多任务Lasso')
plt.legend(loc='upper center')
plt.axis('tight')
plt.ylim([-1.1, 1.1])
plt.show()

```
