# legend

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号

red_patch = mpatches.Patch(color='blue', label='蓝色')
plt.legend(handles=[red_patch])

plt.show()
```

```python
import matplotlib.lines as mlines
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号

blue_line = mlines.Line2D([], [], color='red', marker='*',
                          markersize=15, label='红色星星')
plt.legend(handles=[blue_line])

plt.show()
```

```python
plt.subplot(211)
plt.plot([1, 3, 2], label="线段1")
plt.plot([2, 1, 3], label="线段2")

# bbox_to_anchor 显示的是比例
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
           ncol=2, mode="expand", borderaxespad=0.)

plt.subplot(223)
plt.plot([1, 2, 3], label="线段1")
plt.plot([3, 2, 1], label="线段2")
# Place a legend to the right of this smaller subplot.
plt.legend(bbox_to_anchor=(-0.5, 1), loc='upper left', borderaxespad=0.)

plt.show()
```

# Multiple legends

```python
line1, = plt.plot([1, 3, 2], label="线段A", linestyle='-')
line2, = plt.plot([2,1,3], label="线段B", linestyle='--')

# 为图像创建第一个legend
first_legend = plt.legend(handles=[line1], loc='upper right')

# 手动添加到图像中
plt.gca().add_artist(first_legend)

# 为图像创建第二个legend
plt.legend(handles=[line2], loc='lower right')

plt.show()


```
