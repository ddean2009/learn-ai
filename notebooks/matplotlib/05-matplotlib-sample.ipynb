# 曲线

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号

# 构建数据
t = np.arange(0.0, 2.0, 0.01)
s = 1 + t*t*t

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='x轴', ylabel='y轴',
       title='3次方曲线')
ax.grid()

fig.savefig("test.png")
plt.show()
```

# 多图

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.sin(2 * np.pi * x1) * np.exp(-x1)
y2 = np.sin(2 * np.pi * x2)

fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('两个子图的图形')

ax1.plot(x1, y1, 'or-')
ax1.set_ylabel('图一Y轴')

ax2.plot(x2, y2, '.b-')
ax2.set_xlabel('图二X轴')
ax2.set_ylabel('图二Y轴')

plt.show()
```

# 直方图

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号

np.random.seed(1234567)

# example data
mu = 100  # mean of distribution
sigma = 10  # standard deviation of distribution
x = mu + sigma * np.random.randn(500)

num_bins = 30

fig, ax = plt.subplots()

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, density=True)

# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax.plot(bins, y, '--')
ax.set_xlabel('值')
ax.set_ylabel('概率')
ax.set_title(r'直方图: $\mu=100$, $\sigma=10$')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
```

柱状图

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号


labels = ['A', 'B', 'C', 'D', 'E']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='男性',align='center')
rects2 = ax.bar(x + width/2, women_means, width, label='女性',align='center')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('得分')
ax.set_title('男女得分')
ax.set_xticks(x)
ax.set_xticklabels(labels)

# ax.set_yticks(x)
# ax.set_yticklabels(labels)
# ax.invert_yaxis()

ax.legend()

fig.tight_layout()

plt.show()
```

# 饼图

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号

# 逆时针排序
labels = '初级', '中级', '高级', 'VIP'
sizes = [30, 15, 10, 45]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
